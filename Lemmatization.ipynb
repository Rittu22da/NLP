## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatization = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
'''
POS:-
Noun - 'n'
verb - 'v'
adjective - 'a'
adverb - 'r'
'''
lemmatization.lemmatize('going',pos='v')
```

```python


lemmatization.lemmatize('going', pos='v'),lemmatization.lemmatize('better', pos='a'),lemmatization.lemmatize('cars', pos='n'),lemmatization.lemmatize('running', pos='v')


```
