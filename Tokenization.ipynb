```python
corpus = """ Hey , I am Ritesh .Friends like to call me Rittu . I completed my 7th sem of B.tech . Now i am looking for an internship in an recognized Company.
"""
```

```python
corpus
```

```python
print(corpus)
```

```python
## Tokenization 
## Sentence Tokenizer

from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
documents = sent_tokenize(corpus)
```

```python
documents
```

```python
print(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization 

# Paragraph --> Word
# Sentence ---> Word

from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
